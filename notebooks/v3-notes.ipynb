```python
import numpy as np
import matplotlib.pyplot as plt

# Possible prices demarcated by discrete ticks
# A range is [lower tick, upper tick] aka [i_l, i_u]

# p(i) = 1.0001^i
```

```python
# integer index
# 0, 1, 2, 3, 4...
i = np.arange(0, 100_000, 1)
# prices at every tick
prices = np.power(1.0001, i)
# every price is +- 0.01% from its neighbors
prices[0:10]
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# plt.xticks(np.arange(min(x), max(x)+1, 1.0))
# plt.yticks(np.arange(min(y), max(y)+1, 0.1))

# plot the function
plt.plot(i, prices)

# show the plot
plt.show()
```
