```python
import matplotlib
import matplotlib.pyplot as plt
import math
```

```python
import numpy as np

# x: d
x = np.linspace(0,12,1000)
# impermanent loss formula (d is price change in % of token0)
y = 2*np.sqrt(x)/(1+x) - 1
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

plt.xticks(np.arange(min(x), max(x)+1, 1.0))
plt.yticks(np.arange(min(y), max(y)+1, 0.1))

# plot the function
plt.plot(x,y)

# show the plot
plt.show()
```
